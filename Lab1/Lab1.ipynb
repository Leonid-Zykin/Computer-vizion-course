```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Загрузка изображения
image = cv2.imread('LETI.jpg') 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Построение гистограммы
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

plt.figure()
plt.title("Гистограмма изображения")
plt.xlabel("Интенсивность")
plt.ylabel("Частота")
plt.plot(hist)
plt.xlim([0, 256])
plt.show()

# Профиль изображения (по центру)
profile_x = gray[gray.shape[0] // 2, :]
profile_y = gray[:, gray.shape[1] // 2]

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(profile_x)
plt.title("Профиль по горизонтали (центральная строка)")

plt.subplot(2, 1, 2)
plt.plot(profile_y)
plt.title("Профиль по вертикали (центральный столбец)")
plt.show()

# Горизонтальная и вертикальная проекции
projection_x = np.sum(gray, axis=0)
projection_y = np.sum(gray, axis=1)

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(projection_x)
plt.title("Горизонтальная проекция")

plt.subplot(2, 1, 2)
plt.plot(projection_y)
plt.title("Вертикальная проекция")
plt.show()
```
